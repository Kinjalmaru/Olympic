```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
Olympians = pd.read_csv('../Raw_Data/athlete_events.csv')
Olympians.head()
```

```python
Summer_Olympics = Olympians[Olympians['Season']=='Summer']
Summer_Olympics.head()
```

```python
Best20_Teams = ['United States', 'China', 'Japan', 'Great Britain', 'Russia', 'Australia', 'Netherlands', 'France', 'Germany',
                'Italy', 'Canada', 'Brazil', 'New Zealand', 'Cuba', 'Hungary', 'South Korea', 'Poland', 'Czech Republic',
               'Kenya', 'Norway']
Summer_20 = Summer_Olympics.loc[Summer_Olympics['Team'].isin(Best20_Teams)]
Summer_20.head()
```

```python
Teams_Medals = Summer_20[['Sex','Age', 'Team', 'Sport', 'Year', 'Medal']]
bins= [0,10,20,30,40,50]
labels = ['10','20','30','40','50']
Teams_Medals['Age Group'] = pd.cut(Teams_Medals['Age'], bins=bins, labels=labels, right=False)

Teams_Medals
```

```python
Teams = Teams_Medals.groupby(['Team', 'Sport','Age Group'])['Medal'].count().reset_index()
age_team = Teams.groupby(['Team', 'Sport', 'Medal'])['Age Group'].count().reset_index()
age_team
```
