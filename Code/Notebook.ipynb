```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Olympians = pd.read_csv('../Raw_Data/athlete_events.csv')
Olympians.head()
```

```python
Summer_Olympics = Olympians[Olympians['Season']=='Summer']
Summer_Olympics
```

```python
Teams_Medals = Summer_Olympics[['Sex', 'Team', 'Sport', 'Year', 'Medal']]
Teams_Medals = Teams_Medals[Teams_Medals['Sex']=='M']
Teams_Medals.head()
```

```python
Male_Teams = Teams_Medals[(Teams_Medals["Sport"] == 'Basketball') & (Teams_Medals["Team"] == "United States")]
Male_Teams = Male_Teams[["Team", "Sport", "Year", "Medal"]]
Male_Teams["Medal"] = Male_Teams["Medal"].fillna(0)
Male_Teams
```

```python
               
Male_Teams = Male_Teams.replace("Gold", 0)
Male_Teams = Male_Teams.replace("Silver", 0)
Male_Teams = Male_Teams.replace("Bronze", 1)
Male_Teams = Male_Teams.replace("United States", 0)
Male_Teams = Male_Teams.replace("Basketball", 0)

Male_Teams = Male_Teams.dropna()

Male_Teams
```

```python
X = Male_Teams[['Team', "Sport", "Year"]]  
y = Male_Teams['Medal'] 
print(X.shape, y.shape)
```

```python
Teams_Medals["Sport"].unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
print('Basketball Men')
print(f"Training Data Score: {classifier.score(X_train, y_train)}") 
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
